# Handling Missing Values

**Task 1: Load a Dataset with Missing Values**

Write a Python program to load a dataset with missing values. For this experiment, let's use the Titanic dataset, which contains missing values. If you haven't downloaded it yet, you can get it from this repository.

```python
import pandas as pd

# Load the Titanic dataset
df = pd.read_csv('titanic.csv')
```

```python
# Display the first few rows of the dataset
print("First 5 Rows:\n")
df.head()
```

```python
# Display information about the dataset
print("\nDataset Info:\n")
df.info()
```

```python
# Display the number of missing values in each column
print("\nMissing Values in Each Column:\n")
df.isnull().sum()
```

---

**Task 2: Remove Rows with Missing Values**

```python
# Remove rows with missing values
df_dropped = df.dropna()
```

```python
# Display the first few rows of the cleaned dataset
print("First 5 Rows After Dropping Missing Values:\n")
df_dropped.head()
```

```python
# Display information about the cleaned dataset
print("\nCleaned Dataset Info:\n")
df_dropped.info()
```

---

**Task 3: Fill Missing Values with a Specific Value**

```python
# Fill missing values in the 'Age' column with the mean age
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Fill missing values in the 'Embarked' column with the mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Display the first few rows of the dataset after filling missing values
print("First 5 Rows After Filling Missing Values:\n")
df.head()
```

```python
# Display the number of missing values in each column after filling
print("\nMissing Values in Each Column After Filling:\n")
df.isnull().sum()
```

---

**Task 4: Fill Missing Values Using Forward and Backward Fill**

```python
# Reintroduce missing values for demonstration
df.loc[5:10, 'Age'] = None

# Fill missing values using forward fill
df_ffill = df.fillna(method='ffill')

# Display the first few rows of the datasets after forward fill
print("First 5 Rows After Forward Fill:\n")
df_ffill.head(10)
```

```python
# Fill missing values using backward fill
df_bfill = df.fillna(method='bfill')

# Display the first few rows of the datasets after backward fill
print("\nFirst 5 Rows After Backward Fill:\n")
df_bfill.head(10)
```

---

**Task 5: Interpolate Missing Values**

```python
# Reintroduce missing values for demonstration
df.loc[5:10, 'Age'] = None

# Interpolate missing values
df_interpolated = df.interpolate()

# Display the first few rows of the dataset after interpolation
print("First 5 Rows After Interpolation:\n")
df_interpolated.head(10)
```
